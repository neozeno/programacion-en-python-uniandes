## Creación de subplots

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 8), sharex=True, sharey=True)

# axes es un matriz 2x3, la aplanamos para iterar más fácilmente
axes_flat = axes.flatten()

for i, (ciudad, meses) in enumerate(temperaturas_medias_mensuales.items()):
    ax = axes_flat[i]

    ax.plot(list(meses.keys()), list(meses.values()), "r", label=ciudad)
    ax.set_title(ciudad.capitalize())
    ax.set_xlabel("Mes")
    ax.set_ylabel("Temperatura (°C)")
    ax.set_xticklabels(list(meses.keys()))

plt.tight_layout()
fig.show()
```

> Vale la pena notar en esta imagen que los ejes son compartidos en ​las filas y las columnas, es decir, que en la fila superior no aparecen las etiquetas ​del eje `x` porque se encuentran en la fila de abajo. ​Lo mismo pasa con las etiquetas del eje `y` que se encuentran sólo a la izquierda ​de la figura. ​Además de ahorrar un poco de espacio, esto refuerza la idea de que las seis gráficas ​están presentadas en la misma escala y se pueden comparar con facilidad entre si. Lo que si es diferente en cada gráfica es el título, ​cada una tiene el nombre de la ciudad que le corresponde.

Revisemos ahora el codigo utilizado para la creacion de la grafica anterior. En primer lugar, ​veamos que ahora estamos usando la función `subplots()` en lugar de la función `figure()`. Esta función nos retorna una dupla con 2 cosas: la primera es la figura misma, ​la segunda es una lista o una matriz con los objetos de tipo `ax` ​en los que van a quedar nuestras gráficas. ​En este caso, le indicamos a la función que queriamos organizar las gráficas en dos ​filas y tres columnas, así que en la variable axes nos va a quedar una matriz. ​Si alguna de las dos dimensiones hubiera sido 1, el resultado sería una lista. ​¿Recuerdan que en la leccion anterior teníamos que calcular la posición y el tamaño de ​cada cosa cuando invocábamos a la función `add_axes()`? ​Eso you no lo tenemos que hacer. ​La función `subplots()` se encargará de todos los cálculos. ​Los otros parámetros de estos llamados son `share_x` y `share_y`, ​los cuales nos sirven para indicarle a Matplotlib que queremos que se ​compartan los ejes en las filas y en las columnas. 

Lo que viene en las siguientes líneas es simplemente un recorrido sobre la matriz ​de `axes` armando una gráfica independiente para cada ciudad. `axes.flatten()` convierte la matriz 2x3 en una lista plana de 6 axes, lo que permite usar `enumerate()` una sola vez en lugar de indices `[fila][columna]`. `enumerate()` entrega el índice `i` para acceder al axe correspondiente y el par `(ciudad, meses)` del diccionario simultáneamente. Asi mismo, cambiamos los datos que utilizamos en la función `plot()` ​y el nombre de la ciudad que utilizamos como título de cada gráfica.

Por último, tenemos una invocación a la función `tight_layout()`. ​Esta invocación es necesaria para que Matplotlib recalcule las posiciones de las ​gráficas, teniendo también en cuenta el tamaño de las etiquetas. ​Si no hacemos esto, ​corremos el riesgo de que las gráficas queden montadas una sobre la otra.

## Subplots con dos series

Para mostrar que cada uno de los objetos `ax` que se encuentran en la matriz se comportan exactamente igual que si estuvieran en su propia figura, acá hemos agregado un gemelo para mostrar también las precipitaciones medias mensuales.

```python
precipitacion_min = min(
    v
    for meses in precipitaciones_medias_mensuales.values()
    for v in meses.values()
)

precipitacion_max = max(
    v
    for meses in precipitaciones_medias_mensuales.values()
    for v in meses.values()
)

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 8), sharex=True, sharey=True)
axes_flat = axes.flatten()

for i, (ciudad, meses) in enumerate(temperaturas_medias_mensuales.items()):
    ax = axes_flat[i]

    ax.plot(
        list(meses.keys()),
        list(meses.values()),
        color="steelblue",
        label="Temperatura"
    )

    ax.set_title(ciudad.capitalize())
    ax.set_xlabel("Mes")
    ax.set_ylabel("Temperatura (°C)", color="steelblue")

    ax.tick_params(axis="x", rotation=45)
    ax.tick_params(axis="y", colors="steelblue")
    ax.set_xticklabels(list(meses.keys()))

    ax2 = ax.twinx()
    precipitaciones = precipitaciones_medias_mensuales[ciudad]

    ax2.plot(
        list(precipitaciones.keys()),
        list(precipitaciones.values()),
        color="cornflowerblue",
        label="Precipitación",
        linestyle="--"
   )

    ax2.set_ylabel("Precipitación (mm)", color="cornflowerblue")
    ax2.tick_params(axis="y", colors="cornflowerblue")
    ax2.set_ylim(precipitacion_min, precipitacion_max)

    lines1, labels1 = ax.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax.legend(lines1 + lines2, labels1 + labels2, fontsize="small")

plt.tight_layout()
fig.show()
```

El resultado es que ahora cada gráfica tiene 2 series lineas.

## Subplots y axes

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10), sharex=True, sharey=True)

# Ajustamos el espacio superior para darle lugar al ax superior
fig.subplots_adjust(top=0.72)

# Gráfica de dispersión encima: [left, bottom, width, height] en fracción de la figura
ax_scatter = fig.add_axes([0.1, 0.78, 0.8, 0.18])

for ciudad in temperaturas_medias_mensuales:
    temps = list(temperaturas_medias_mensuales[ciudad].values())
    precips = list(precipitaciones_medias_mensuales[ciudad].values())
    ax_scatter.scatter(temps, precips, label=ciudad.capitalize(), alpha=0.7, s=40)

ax_scatter.set_xlabel("Temperatura media mensual (°C)")
ax_scatter.set_ylabel("Precipitación (mm)")
ax_scatter.set_title("Temperatura vs Precipitación — todas las ciudades")
ax_scatter.legend(fontsize=7, ncol=3)

# Subplots de temperatura
precipitacion_min = min(
    v
    for meses in precipitaciones_medias_mensuales.values()
    for v in meses.values()
)
precipitacion_max = max(
    v
    for meses in precipitaciones_medias_mensuales.values()
    for v in meses.values()
)

axes_flat = axes.flatten()

for i, (ciudad, meses) in enumerate(temperaturas_medias_mensuales.items()):
    ax = axes_flat[i]

    ax.plot(
        list(meses.keys()),
        list(meses.values()),
        color="steelblue",
        label="Temp. mensual",
        marker="o",
        markersize=4,
    )
    ax.set_title(ciudad.capitalize())
    ax.set_xlabel("Mes")
    ax.set_ylabel("Temperatura (°C)", color="steelblue")
    ax.tick_params(axis="y", labelcolor="steelblue")
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right")

    ax2 = ax.twinx()
    precipitaciones = precipitaciones_medias_mensuales[ciudad]
    ax2.plot(
        list(precipitaciones.keys()),
        list(precipitaciones.values()),
        color="cornflowerblue",
        label="Precipitación",
        marker="o",
        markersize=4,
        linestyle="--",
    )
    ax2.set_ylabel("Precipitación (mm)", color="cornflowerblue")
    ax2.tick_params(axis="y", labelcolor="cornflowerblue")
    ax2.set_ylim(precipitacion_min, precipitacion_max)

    lines1, labels1 = ax.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax.legend(lines1 + lines2, labels1 + labels2, fontsize=7)

plt.show()
```
